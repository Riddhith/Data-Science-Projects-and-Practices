```python
all_data=pd.read_csv('all_months_data.csv')
```

```python
all_data[['Order_Date','Order_Time']] = all_data['Order Date'].str.split(' ',expand=True)
```

```python
all_data.head()
```

```python
all_data = all_data.dropna(how='all')
```

```python
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

```python
all_data.drop(['Order_Date','Order_Time'],axis=1)
```

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

#### Q1. What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,results)
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Month Number")
plt.show()
```

#### Q2. What city has the highest number of sales?

```python
all_data.head(10)
```

```python
# all_data[['Street','City', 'Code']] = all_data['Purchase Address'].str.split(',', expand=True)
def get_city(address):
    return address.split(",")[1]
def get_state(address):
    return address.split(',')[2][1:3]
# all_data = all_data.drop(['Column'],axis=1)
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) +" "+ get_state(x))
```

```python
all_data
```

```python
results = all_data.groupby('City').sum()
```

```python
# import matplotlib.pyplot as plt
cities = [city for city,df in all_data.groupby('City')]
plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel("Sales in USD ($)")
plt.xlabel("City Name")
plt.show()
```

#### Q4. What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df['Grouped'] = df.groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))
df
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
count = Counter()
for list in df['Grouped']:
    sublist = list.split(',')
    count.update(Counter(combinations(sublist,2)))
for key,value in count.most_common(10):
    print(key,value)
```

```python
# Select only numerical columns from the DataFrame
numerical_columns = all_data.select_dtypes(include=['number'])

# Group by 'Product' and then sum only the numerical columns
product_group = all_data.groupby('Product')

# Display the result
quantity_ordered = product_group[numerical_columns.columns].sum()['Quantity Ordered']

products = [product for product,df in product_group]
plt.bar(products,quantity_ordered)
plt.xticks(products,rotation='vertical',size=8)
plt.ylabel("Quantity Ordered")
plt.xlabel("Products")
plt.show()
```

```python
prices = all_data.groupby('Product')['Price Each'].mean()
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered,color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Products')
ax1.set_xticklabels(products,rotation='vertical',size=8)
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Prices', color='b')

plt.show()
```
